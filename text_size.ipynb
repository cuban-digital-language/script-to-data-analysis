```python
from submodulos.data import get_all_text_with_details 
from submodulos.data.telegram import get_details_text
from tools import get_progressbar, groupBy
import os
import numpy as np
import matplotlib.pyplot as plt

texts = get_all_text_with_details(f'{os.getcwd()}/submodulos/data/')
```

```python
text_list = {
    'CubaDebate': [],
    'Facebook': [],
    'Twitter': [],
    'Telegram': []
}

for t, d, n in texts:
    text_list[n].append(t)

```

```python
def size_analyze(_list):
    _len = len(_list) 
    _distribution = [len(t) for t in _list]
    _mean = sum(_distribution)/_len
    return _len, _distribution, _mean


def histogram(data, bins): 

    plt.xlim([min(data)-5, max(data)+5])

    plt.hist(data, bins=bins, alpha=0.5)
    plt.title('Random Gaussian data (fixed bin size)')
    plt.xlabel('variable X (bin size = 5)')
    plt.ylabel('count')

    plt.show()
```

```python
l, d, m = size_analyze(text_list['Telegram'])
mn, mx =min(d), max(d)
print('text size', l)
print('text mean length', m, mn, mx)


bins = np.arange(mn, mx, m) # fixed bin size

histogram(d, bins)
```

```python
l, d, m = size_analyze(text_list['Twitter'])
mn, mx =min(d), max(d)

print('text size', l)
print('text mean length', m, min(d), max(d))
bins = np.arange(mn, mx, m) # fixed bin size


histogram(d, bins)
```

```python
l, d, m = size_analyze([t for t, _, _ in texts])
mn, mx =min(d), max(d)

print('text size', l)
print('text mean length', m, min(d), max(d))
bins = np.arange(mn, mx, m) # fixed bin size


histogram(d, bins)
```
